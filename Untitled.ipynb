```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('Churn_Modelling.csv')
```

```python
df.head()
```

```python
df_1 = df.drop(['RowNumber','Surname','CustomerId'],axis=1)
```

```python
df_1.head()
```

```python
df_1.isnull().sum()
```

```python
df_1.dropna(inplace=True)
```

```python
df_1.isnull().sum()
```

```python
df_1.head()
```

```python
df_1.describe()
```

```python
df_1 = pd.get_dummies(df_1,columns=['Geography','Gender'],drop_first=True)
```

```python
df_1.head()
```

```python
df_1['Geography_Germany']=df_1['Geography_Germany'].astype(int)
```

```python
df_1['Geography_Spain'] = df_1['Geography_Spain'].astype(int)
```

```python
df_1['Gender_Male'] = df_1['Gender_Male'].astype(int)
```

```python
df_1.head()
```

```python
x = df_1.drop('Exited',axis=1)
```

```python
y = df_1['Exited']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train_std = scaler.fit_transform(x_train)
```

```python
x_test_std = scaler.transform(x_test)
```

```python
x_train_std
```

```python
import tensorflow as tf
import tf_keras as tfk
```

```python
from tensorflow import keras
from tensorflow.keras import Sequential 
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(60,activation='relu',input_dim=11))
model.add(Dense(30,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='Adam')
```

```python
model.fit(x_train_std,y_train,epochs=100) # by increasing the no. of epochs we can improve our model accuracy
```

```python
y_pred = model.predict(x_test_std)
```

```python
y_preds = np.where(y_pred>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_preds)
```

```python
model.save('model.h5')
```
